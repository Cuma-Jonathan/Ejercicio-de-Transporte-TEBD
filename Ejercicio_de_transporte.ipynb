```python
import requests
from bs4 import BeautifulSoup
import os
import re

# URL de la página de archivos del Transportation Energy Data Book
url = 'https://tedb.ornl.gov/archive/'

# Realizar una solicitud HTTP para obtener el contenido de la página
response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')

# Buscar todos los enlaces que contienen el texto "Transportation Energy Data Book"
links = soup.find_all('a', string=lambda text: "Transportation Energy Data Book" in text if text else False)

# Explicación: Usamos una expresión regular para identificar ediciones entre 28 a 40
edition_regex = re.compile(r'(\d+)', re.I)

# Directorio temporal en Colab
save_directory = '/content/temporal'  # Cambia esto a tu directorio deseado

# Crear el directorio si no existe
os.makedirs(save_directory, exist_ok=True)

# Iterar sobre los enlaces encontrados
for link in links:
    # Extraer la URL del archivo PDF
    pdf_url = link.get('href')
    # Extraer el nombre del archivo PDF
    pdf_name = pdf_url.split('/')[-1]

    # Obtener el número de edición del PDF
    edition_match = edition_regex.search(pdf_name)
    if edition_match:
        edition_number = int(edition_match.group(1))
        # Verificar si la edición está entre 28 y 40
        if 28 <= edition_number <= 40:
            # Combinar la ruta del directorio con el nombre del archivo PDF
            pdf_path = os.path.join(save_directory, pdf_name)

            # Realizar la solicitud para descargar el archivo PDF
            pdf_response = requests.get(pdf_url)

            # Guardar el archivo PDF en el directorio especificado
            with open(pdf_path, 'wb') as file:
                file.write(pdf_response.content)

            print(f'Archivo Descargado: {pdf_path}')

print('Descarga completada.')
```

```python
import os
!pip install PyMuPDF
import fitz  # PyMuPDF
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from io import StringIO

# Directorio temporal en Colab
temporal_directory = '/content/temporal'

# Obtener la lista de archivos en el directorio "temporal"
pdf_files = [file for file in os.listdir(temporal_directory) if file.endswith('.pdf')]

# Configuración del estilo de seaborn para mejorar la estética del gráfico
sns.set(style="whitegrid")

# Iterar sobre cada archivo PDF en el directorio temporal
for pdf_file in pdf_files:
    pdf_path = os.path.join(temporal_directory, pdf_file)

    # Resto del código para abrir y procesar el archivo PDF según tus necesidades
    pdf_document = fitz.open(pdf_path)
    table_page = None

    # Informar que se está procesando un reporte específico
    print(f'Procesando reporte: {pdf_file}')

    # Iterar sobre las páginas del documento PDF
    for page_number in range(len(pdf_document)):
        page = pdf_document.load_page(page_number)
        text = page.get_text()

        # Buscar la página que contiene la tabla deseada
        if "Domestic Consumption of Transportation Energy by Mode and Fuel Type" in text:
            table_page = page
            break

    # Si se encontró la página de la tabla
    if table_page:
        table_content = table_page.get_text()

        # Encontrar los índices de inicio y fin de la tabla
        start_index = table_content.find("Gasoline")
        end_index = table_content.find("Source") + len("Source")

        # Extraer la porción deseada de texto
        desired_text = table_content[start_index:end_index]

        # Preprocesar el texto para facilitar su análisis
        lines = desired_text.split('\n')
        clean_lines = [line.strip() for line in lines if line.strip() != '']

        # Inicializar listas para categorías y datos
        categories = []
        current_category = ''
        data = []

        # Iterar sobre las líneas procesadas
        for line in clean_lines:
            # Verificar si la línea contiene datos numéricos o guiones (líneas de datos)
            if line.replace('.', '', 1).isdigit() or '-' in line:
                data.append(current_category + ',' + line)
            else:
                current_category = line  # Asumir que las líneas sin datos son categorías

        # Unir las líneas de datos en una cadena simulando un objeto de archivo
        data_str = '\n'.join(data)

        # Convertir la cadena formateada en un DataFrame de pandas
        df = pd.read_csv(StringIO(data_str), header=None, names=['Mode', 'Fuel Type', 'Gasoline', 'Diesel fuel', 'Liquefied petroleum gas', 'Jet fuel', 'Residual fuel oil', 'Natural gas', 'Electricity', 'Total'])

        # Reemplazar los valores 'NA' con 0 en el DataFrame
        df = df.fillna(0)

        # Imprimir el DataFrame
        print(df.to_string(index=False))

        # Visualizar los datos con un gráfico de barras
        plt.figure(figsize=(12, 8))

        # Derretir el DataFrame al formato largo para facilitar la representación gráfica
        df_long = df.melt(id_vars=['Mode', 'Fuel Type'], var_name='Energy Source', value_name='Consumption')

        # Gráfico de barras
        sns.barplot(data=df_long, x='Consumption', y='Energy Source', hue='Mode', ci=None)
        plt.title('Domestic Consumption of Transportation Energy by Mode and Fuel Type')
        plt.xlabel('Consumption')
        plt.ylabel('Energy Source')
        plt.tight_layout()
        plt.show()

    # Mensaje en caso de que la tabla no se encuentre en el PDF
    else:
        print(f"Table not found in the PDF: {pdf_path}")

    # Imprimir mensaje indicando que el procesamiento ha sido completado para el reporte actual
    print(f'Procesado: {pdf_path}')

# Mensaje al finalizar el procesamiento de todos los archivos PDF
print('Procesamiento listo.')
```
